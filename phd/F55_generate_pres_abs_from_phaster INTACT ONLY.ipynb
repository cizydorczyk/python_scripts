```python
# Read in cdhit-est clusters file:
clusters = {}
with open("/home/conrad/ecoli_all/PHASTER_TESTING/intact_only_phages/ip99-98.clstr", 'r') as infile1:
    cluster_block = []
    cluster_header = ""
    
    for line in infile1:
        if line.startswith(">"):
            if cluster_header:
                prev_header = cluster_header
                cluster_header = line.strip()[1:]
            else:
                cluster_header = line.strip()[1:]

            if cluster_block:
                clusters[prev_header] = cluster_block # set entries for existing cluster
                cluster_block = [] # reset cluster block for new cluster block entries
#             clusters[cluster_header] = [] # create dict entry for new cluster;
#                                             no need to initialize dict entry b/c not interatively adding to it
        else:
#             cluster_block.append(line.strip().split("\t")[1].split(" ")[1][1:-3]) # add entries to existing cluster
            cluster_block.append(line.strip().split("\t")[1]) # add entries to existing cluster

    if cluster_block:
        clusters[cluster_header] = cluster_block # for final cluster entry

print(clusters)
```

```python
# Load isolates list:
with open("/home/conrad/ecoli_all/ec_isolate_list.txt", "r") as infile2:
    isolates = [line.strip() for line in infile2]

print(isolates)
```

```python
isolate_phage_dict = {}
isolate_phage_dict_binary = {}
for isolate in isolates:
    isolate_phages = []
    isolate_phages_binary = []
    
    for cluster in clusters:
        cluster_entries = clusters[cluster]
        if any(isolate in entry for entry in cluster_entries):
            isolate_phages.append([entry for entry in cluster_entries if isolate in entry][0]) # hard code [0]
                                                                # b/c only should ever be 1 matching entry
            isolate_phages_binary.append("1")
        else:
            isolate_phages.append("")
            isolate_phages_binary.append("0")
    
    isolate_phage_dict[isolate] = isolate_phages
    isolate_phage_dict_binary[isolate] = isolate_phages_binary

```

```python
import pandas as pd

output_df = pd.DataFrame.from_dict(isolate_phage_dict, orient="index")
output_df
```

```python
# Write df to file:
output_df.to_csv(path_or_buf = "/home/conrad/ecoli_all/PHASTER_TESTING/intact_only_phages/phage_pres_abs.txt", sep="\t")
```
